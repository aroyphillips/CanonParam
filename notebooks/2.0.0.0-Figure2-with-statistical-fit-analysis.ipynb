```python
open_fooof_armse = fooof_rmse_df[[col for col in fooof_rmse_df.columns if 'open' in col]].mean(axis=1)
closed_fooof_armse = fooof_rmse_df[[col for col in fooof_rmse_df.columns if 'closed' in col]].mean(axis=1)
open_our_armse = our_rmse_df[[col for col in our_rmse_df.columns if 'open' in col]].mean(axis=1)
closed_our_armse = our_rmse_df[[col for col in our_rmse_df.columns if 'closed' in col]].mean(axis=1)

fig, axs = plt.subplots(1,2, figsize=(12,4))
axs[0].hist(open_fooof_armse, bins=20, alpha=0.5, label='SpecParam')
axs[0].hist(open_our_armse, bins=20, alpha=0.5, label='Our Method')
axs[0].set_title('Open Eyes RMSE')
axs[0].set_xlabel('RMSE')
axs[0].set_ylabel('Number of Subjects')
axs[0].legend()
axs[1].hist(closed_fooof_armse, bins=20, alpha=0.5, label='SpecParam')
axs[1].hist(closed_our_armse, bins=20, alpha=0.5, label='Our Method')
axs[1].set_title('Closed Eyes RMSE')
axs[1].set_xlabel('RMSE')
axs[1].set_ylabel('Number of Subjects')
axs[1].legend()

plt.show()

open_wilcoxon = wilcoxon(open_fooof_armse, open_our_armse, alternative='greater')
closed_wilcoxon = wilcoxon(closed_fooof_armse, closed_our_armse, alternative='greater')

# holm-bonferroni correction
print(open_wilcoxon)
print(closed_wilcoxon)
from statsmodels.stats.multitest import multipletests
adjusted_pvals = multipletests([open_wilcoxon.pvalue, closed_wilcoxon.pvalue], method='holm')
print(f"Adjusted p-values: {adjusted_pvals[1]}")


open_ttest = ttest_rel(open_fooof_armse, open_our_armse, alternative='greater')
closed_ttest = ttest_rel(closed_fooof_armse, closed_our_armse, alternative='greater')
print(open_ttest)
print(closed_ttest)
adjusted_tpvals = multipletests([open_ttest.pvalue, closed_ttest.pvalue], method='holm')
print(f"Adjusted p-values: {adjusted_tpvals[1]}")
```

```python
channels = list(np.unique([col.split('_')[0] for col in our_rsq_df.columns]))

rmse_wilcoxon_pvals = np.empty((len(channels), 2))
rmse_ttest_pvals = np.empty((len(channels), 2))
rsq_wilcoxon_pvals = np.empty((len(channels), 2))
rsq_ttest_pvals = np.empty((len(channels), 2))
for cdx, channel in enumerate(channels):
    for sdx, state in enumerate(['open', 'closed']):
        rmse_fooof = fooof_rmse_df[[col for col in fooof_rmse_df.columns if channel in col and state in col]].mean(axis=1)
        rmse_our = our_rmse_df[[col for col in our_rmse_df.columns if channel in col and state in col]].mean(axis=1)
        rsq_fooof = fooof_rsq_df[[col for col in fooof_rsq_df.columns if channel in col and state in col]].mean(axis=1)
        rsq_our = our_rsq_df[[col for col in our_rsq_df.columns if channel in col and state in col]].mean(axis=1)
        rmse_wilcoxon_pvals[cdx, sdx] = wilcoxon(rmse_fooof, rmse_our, alternative='greater').pvalue
        rmse_ttest_pvals[cdx, sdx] = ttest_rel(rmse_fooof, rmse_our, alternative='greater').pvalue
        rsq_wilcoxon_pvals[cdx, sdx] = wilcoxon(rsq_fooof, rsq_our, alternative='less').pvalue
        rsq_ttest_pvals[cdx, sdx] = ttest_rel(rsq_fooof, rsq_our, alternative='less').pvalue

print(f"Min and max RMSE Wilcoxon p-values: {np.min(rmse_wilcoxon_pvals)} and {np.max(rmse_wilcoxon_pvals)}")
print(f"Min and max RMSE T-test p-values: {np.min(rmse_ttest_pvals)} and {np.max(rmse_ttest_pvals)}")
print(f"Min and max R-Squared Wilcoxon p-values: {np.min(rsq_wilcoxon_pvals)} and {np.max(rsq_wilcoxon_pvals)}")
print(f"Min and max R-Squared T-test p-values: {np.min(rsq_ttest_pvals)} and {np.max(rsq_ttest_pvals)}")


adjusted_rmse_wilcoxon_pvals = multipletests(rmse_wilcoxon_pvals.flatten(), method='holm')[1].reshape(rmse_wilcoxon_pvals.shape)
adjusted_rmse_ttest_pvals = multipletests(rmse_ttest_pvals.flatten(), method='holm')[1].reshape(rmse_ttest_pvals.shape)
adjusted_rsq_wilcoxon_pvals = multipletests(rsq_wilcoxon_pvals.flatten(), method='holm')[1].reshape(rsq_wilcoxon_pvals.shape)
adjusted_rsq_ttest_pvals = multipletests(rsq_ttest_pvals.flatten(), method='holm')[1].reshape(rsq_ttest_pvals.shape)

print(f"Adjusted Min and max RMSE Wilcoxon p-values: {np.min(adjusted_rmse_wilcoxon_pvals)} and {np.max(adjusted_rmse_wilcoxon_pvals)}")
print(f"Adjusted Min and max RMSE T-test p-values: {np.min(adjusted_rmse_ttest_pvals)} and {np.max(adjusted_rmse_ttest_pvals)}")
print(f"Adjusted Min and max R-Squared Wilcoxon p-values: {np.min(adjusted_rsq_wilcoxon_pvals)} and {np.max(adjusted_rsq_wilcoxon_pvals)}")
print(f"Adjusted Min and max R-Squared T-test p-values: {np.min(adjusted_rsq_ttest_pvals)} and {np.max(adjusted_rsq_ttest_pvals)}")
```

```python
all_wilcoxon_pval_df = pd.DataFrame(np.concatenate([rmse_wilcoxon_pvals, rsq_wilcoxon_pvals], axis=1), columns=['Open RMSE Wilcoxon', 'Closed RMSE Wilcoxon', 'Open R-Squared Wilcoxon', 'Closed R-Squared Wilcoxon'], index=channels)

all_ttest_pval_df = pd.DataFrame(np.concatenate([rmse_ttest_pvals, rsq_ttest_pvals], axis=1), columns=['Open RMSE T-test', 'Closed RMSE T-test', 'Open R-Squared T-test', 'Closed R-Squared T-test'], index=channels)
adjusted_all_wilcoxon_pval_df = pd.DataFrame(np.concatenate([adjusted_rmse_wilcoxon_pvals, adjusted_rsq_wilcoxon_pvals], axis=1), columns=['Open RMSE Holm-Bonferroni Adjusted Wilcoxon Signed Rank p-val', 'Closed RMSE Holm-Bonferroni Adjusted Wilcoxon Signed Rank p-val', 'Open R-Squared Holm-Bonferroni Adjusted Wilcoxon Signed Rank p-val', 'Closed R-Squared Holm-Bonferroni Adjusted Wilcoxon Signed Rank p-val'], index=channels)
adjusted_all_ttest_pval_df = pd.DataFrame(np.concatenate([adjusted_rmse_ttest_pvals, adjusted_rsq_ttest_pvals], axis=1), columns=['Open RMSE T-test', 'Closed RMSE T-test', 'Open R-Squared T-test', 'Closed R-Squared T-test'], index=channels)


adjusted_all_wilcoxon_pval_df
```

# now we go by frequency

```python
max_freqs = [1.5, 4, 8, 12.5, 30, 70, 150, 250]
all_freq_rmse = np.ones((len(subjs), len(max_freqs), len(channels)*2, 2))*-1000 # subjs, max freqs, channels *  open/closed, us/foooof
all_freq_rsq = np.ones((len(subjs), len(max_freqs), len(channels)*2, 2))*-1000# subjs, max freqs, channels *  open/closed, us/foooof
ap_freq_rmse = np.ones((len(subjs), len(max_freqs), len(channels)*2, 2))*-1000
ap_freq_rsq = np.ones((len(subjs), len(max_freqs), len(channels)*2, 2))*-1000
gaussian_rmse = np.ones((len(subjs),  len(max_freqs),len(channels)*2, 2))*-1000
gaussian_rsq = np.ones((len(subjs),  len(max_freqs), len(channels)*2, 2))*-1000
standard_bands = [(0.3, 1.5), (1.5, 4), (4, 8), (8, 12.5), (12.5, 30), (30, 70), (70, 150)]
all_train_subjs = list(all_parameterized_spectra.keys())
for subjdx, subj in enumerate(all_train_subjs):
    for sdx, state in enumerate(['open', 'closed']):
        psd, freqs = psd_dict[subj][state].get_data(return_freqs=True)
        norm_psd = du.PSDScaler(normalize_method='median', channel_independence=False, normalize_a=None, normalize_b=None).fit_transform(psd)
        processed_freqs, processed_psd = process_power_spectrum(freqs, np.mean(norm_psd, axis=0)[chdx])
        processed_psd = np.log10(processed_psd)
        assert channels == list(psd_dict[subj][state].ch_names)
        for cdx, channel in enumerate(channels):


            our_ap_params = all_parameterized_spectra[subj][state][channel]['aperiodic_params']
            our_gaussian_params = all_parameterized_spectra[subj][state][channel]['gaussian_params']
            our_spectrum, our_peak, our_ap = reconstruct_bandparam_spectrum(processed_freqs, our_ap_params, our_gaussian_params, standard_bands)

            fooof_ap_params = all_fooof_parameterized_spectra[subj][state][channel]['aperiodic_params']
            fooof_gaussian_params = all_fooof_parameterized_spectra[subj][state][channel]['peak_params']
            fooof_spectrum, fooof_peak, fooof_ap = reconstruct_fooof_spectrum(processed_freqs, fooof_ap_params, fooof_gaussian_params)

            for fdx, max_freq in enumerate(max_freqs):
                
                freq_mask = processed_freqs <= max_freq
                fdx_rmse = np.sqrt(np.mean((our_spectrum[freq_mask]-processed_psd[freq_mask])**2))
                fdx_rsq = np.corrcoef(our_spectrum[freq_mask], processed_psd[freq_mask])[0,1]**2 ## this is how FOOOF calculates R^2
                fooof_fdx_rmse = np.sqrt(np.mean((fooof_spectrum[freq_mask]-processed_psd[freq_mask])**2))
                fooof_fdx_rsq = np.corrcoef(fooof_spectrum[freq_mask], processed_psd[freq_mask])[0,1]**2

                fdx_ap_rmse = np.sqrt(np.mean((our_ap[freq_mask]-processed_psd[freq_mask])**2))
                fdx_ap_rsq = np.corrcoef(our_ap[freq_mask], processed_psd[freq_mask])[0,1]**2
                fooof_fdx_ap_rmse = np.sqrt(np.mean((fooof_ap[freq_mask]-processed_psd[freq_mask])**2))
                fooof_fdx_ap_rsq = np.corrcoef(fooof_ap[freq_mask], processed_psd[freq_mask])[0,1]**2

                our_non_ap = our_spectrum - our_ap
                fooof_non_ap = fooof_spectrum - fooof_ap
                our_original_residual = processed_psd - our_ap
                fooof_original_residual = processed_psd - fooof_ap

                fdx_gaussian_rmse = np.sqrt(np.mean((our_non_ap[freq_mask]-our_original_residual[freq_mask])**2))
                fdx_gaussian_rsq = np.corrcoef(our_non_ap[freq_mask], our_original_residual[freq_mask])[0,1]**2
                fooof_fdx_gaussian_rmse = np.sqrt(np.mean((fooof_non_ap[freq_mask]-fooof_original_residual[freq_mask])**2))
                fooof_fdx_gaussian_rsq = np.corrcoef(fooof_non_ap[freq_mask], fooof_original_residual[freq_mask])[0,1]**2


                if np.isnan(fdx_rmse) or np.isnan(fdx_rsq):
                    print(f"Subject {subj} {state} {channel} {cdx} has BandParams NaN values")
                    viz_subj = subj
                    viz_state = state
                    viz_channel = channel
                    
                if np.isnan(fooof_fdx_rmse) or np.isnan(fooof_fdx_rsq):
                    print(f"Subject {subj} {state} {channel} {cdx} has FOOOOF  NaN values")
                    viz_subj = subj
                    viz_state = state
                    viz_channel = channel

                all_freq_rmse[subjdx, fdx, cdx*2+sdx, 0] = fdx_rmse
                all_freq_rmse[subjdx, fdx, cdx*2+sdx, 1] = fooof_fdx_rmse
                all_freq_rsq[subjdx, fdx, cdx*2+sdx, 0] = fdx_rsq
                all_freq_rsq[subjdx, fdx, cdx*2+sdx, 1] = fooof_fdx_rsq
                ap_freq_rmse[subjdx, fdx, cdx*2+sdx, 0] = fdx_ap_rmse
                ap_freq_rmse[subjdx, fdx, cdx*2+sdx, 1] = fooof_fdx_ap_rmse
                ap_freq_rsq[subjdx, fdx, cdx*2+sdx, 0] = fdx_ap_rsq
                ap_freq_rsq[subjdx, fdx, cdx*2+sdx, 1] = fooof_fdx_ap_rsq
                gaussian_rmse[subjdx, fdx, cdx*2+sdx, 0] = fdx_gaussian_rmse
                gaussian_rmse[subjdx, fdx, cdx*2+sdx, 1] = fooof_fdx_gaussian_rmse
                gaussian_rsq[subjdx, fdx, cdx*2+sdx, 0] = fdx_gaussian_rsq
                gaussian_rsq[subjdx, fdx, cdx*2+sdx, 1] = fooof_fdx_gaussian_rsq


print(f"Missed filling in {np.sum(all_rmse_means == -1000)} rmse values")
print(f"Missed filling in {np.sum(all_rsq_means == -1000)} rsq values")

all_freq_rmse_means = np.nanmean(all_freq_rmse)
all_freq_rmse_sems = np.nanstd(all_freq_rmse)/np.sqrt(len(all_freq_rmse.flatten()))
all_freq_rsq_means = np.nanmean(all_freq_rsq)
all_freq_rsq_sems = np.nanstd(all_freq_rsq)/np.sqrt(len(all_freq_rsq.flatten()))

print(f"Average RMSE: {all_freq_rmse_means:.3f} ± {all_freq_rmse_sems:.3f}")
print(f"Average R-Squared: {all_freq_rsq_means:.3f} ± {all_freq_rsq_sems:.3f}")
for fdx, max_freq in enumerate(max_freqs):
    print(f"Average  RMSE up to {max_freq}: {np.nanmean(all_freq_rmse[:, fdx, ...]):.3f} ± {np.nanstd(all_freq_rmse[:, fdx, ...])/np.sqrt(len(all_freq_rmse[:, fdx, ...].flatten())):.3f}")
    print(f"Average  R-Squared up to {max_freq}: {np.nanmean(all_freq_rsq[:, fdx, ...]):.3f} ± {np.nanstd(all_freq_rsq[:, fdx, ...])/np.sqrt(len(all_freq_rsq[:, fdx, ...].flatten())):.3f}")
```
