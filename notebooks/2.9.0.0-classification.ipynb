# unpack fit results

```python
our_param_dfs = out_dict['param_dfs']
all_params = our_param_dfs['all_param']
open_params = all_params[[col for col in all_params.columns if 'open' in col]]
closed_params = all_params[[col for col in all_params.columns if 'closed' in col]]
diff_params = open_params.values - closed_params.values
diff_cols = [col.replace('open', 'diff') for col in open_params.columns]
diff_params = pd.DataFrame(diff_params, columns=diff_cols, index=open_params.index)
```

```python
print(f"fts: {np.unique([col.split('_')[-1] for col in diff_cols])}")
```

```python
diff_features = diff_params[[col for col in diff_params.columns if 'rsq' not in col and 'mse' not in col]]
chois = ['F7', 'F8', 'T3', 'T4']
diffoi_features = diff_features[[col for col in diff_features.columns if col.split('_')[0] in chois]]
diffoi_features
```

# univariate feature comparison

To compute this, we look at the open - closed features and take the mann-whitney u test to see if the distributions are different. We then plot the top three.

```python
diff_feature_mwu_pvals = fu.get_mwup_values(diffoi_features, fu.get_y_from_df(diffoi_features), adjust_method='holm')
diff_feature_mwu_pvals.sort_values(by='p_values', ascending=True)
```
