# establish baseline ML

```python
training_psd_dict = training_dict['psd_dict'] # {subj: {state: epochSpectra}}
standard_bands = [(0.3, 1.5), (1.5, 4), (4, 8), (8, 12.5), (12.5, 30), (30, 70), (70, 150)]
bandnames = ['low delta', 'delta', 'theta', 'alpha', 'beta', 'gamma', 'chi']
def extract_band_powers(epochSpectra, bands, channels):
    assert channels == epochSpectra.ch_names
    band_powers = np.zeros((len(bands), len(channels)))
    freqs = epochSpectra.freqs
    data = epochSpectra.get_data()
    for i, (low, high) in enumerate(bands):
        band_mean = np.mean(data[:, :, (freqs >= low) & (freqs < high)], axis=(0, 2))
        band_powers[i, :] = band_mean
    return band_powers

def extract_all_subj_band_power(subj_dict, bands, channels):
    subj_band_powers = {}
    for subj, states in subj_dict.items():
        subj_band_powers[subj] = {}
        for state, epochSpectra in states.items():
            subj_band_powers[subj][state] = extract_band_powers(epochSpectra, bands, channels)
    return subj_band_powers

def convert_subj_band_power2df(subj_band_powers, bandnames, channels):
    states = subj_band_powers[list(subj_band_powers.keys())[0]]
    out_array = np.empty((len(subj_band_powers), len(states) * len(channels) * len(bandnames)))
    subjs = list(subj_band_powers.keys())
    for i, subj in enumerate(subjs):
        state_bps = subj_band_powers[subj]
        for sdx, state in enumerate(states):
            state_band_power = state_bps[state]
            assert state_band_power.shape[1] == len(channels)
            for cdx, channel in enumerate(channels):
                for bdx, band in enumerate(bandnames):
                    curr_indx = sdx * len(channels) * len(bandnames) + cdx * len(bandnames) + bdx
                    out_array[i, curr_indx] = state_band_power[bdx, cdx]

    col_names=  [f"{channel}_{state}_{band}" for state in states.keys() for channel in channels for band in bandnames]
    return pd.DataFrame(out_array, columns=col_names, index=subj_band_powers.keys())

band_powers = extract_all_subj_band_power(training_psd_dict, standard_bands, all_channels)
band_powers_df = convert_subj_band_power2df(band_powers, bandnames, all_channels)
```

```python
def make_homologous_ref_df(df, all_channels):
    open_params = df[[col for col in df.columns if 'open' in col]]
    closed_params = df[[col for col in df.columns if 'closed' in col]]
    diff_params = open_params.values - closed_params.values
    diff_cols = [col.replace('open', 'diff') for col in open_params.columns]
    diff_params = pd.DataFrame(diff_params, columns=diff_cols, index=open_params.index)
    left_side_params = diff_params[[col for col in diff_params.columns if col.split('_')[0][-1].isdigit() and int(col.split('_')[0][-1]) % 2 == 1]]
    right_side_params = diff_params[[col for col in diff_params.columns if col.split('_')[0][-1].isdigit() and int(col.split('_')[0][-1]) % 2 == 0]]
    homologous_diff_params = np.abs(left_side_params.values - right_side_params.values)
    homologous_diff_cols = [f"|{col.split('_')[0]}-{col.split('_')[0][:-1]}{int(col.split('_')[0][-1])+1}|_{'_'.join(col.split('_')[1:])}" for col in left_side_params.columns]
    homologous_diff_params = pd.DataFrame(homologous_diff_params, columns=homologous_diff_cols, index=left_side_params.index)


    return homologous_diff_params

homologous_band_powers = make_homologous_ref_df(band_powers_df, all_channels)
homologous_band_powers
```

```python
def generate_referenced_homologous_pairs(param_feature_df):
    """
    given a dataframe of spectra parameterizations, generate a dataframe of homologous pairs of features
    """

    open_params = param_feature_df[[col for col in param_feature_df.columns if 'open' in col]]
    closed_params = param_feature_df[[col for col in param_feature_df.columns if 'closed' in col]]
    diff_params = open_params.values - closed_params.values
    diff_cols = [col.replace('open', 'diff') for col in open_params.columns]
    diff_params = pd.DataFrame(diff_params, columns=diff_cols, index=open_params.index)

    left_side_params = diff_params[[col for col in diff_params.columns if col.split('_')[0][-1].isdigit() and int(col.split('_')[0][-1]) % 2 == 1]]
    right_side_params = diff_params[[col for col in diff_params.columns if col.split('_')[0][-1].isdigit() and int(col.split('_')[0][-1]) % 2 == 0]]
    homologous_diff_params = np.abs(left_side_params.values - right_side_params.values)
    homologous_diff_cols = [f"|{col.split('_')[0]}-{col.split('_')[0][:-1]}{int(col.split('_')[0][-1])+1}|_{'_'.join(col.split('_')[1:])}" for col in left_side_params.columns]
    homologous_diff_params = pd.DataFrame(homologous_diff_params, columns=homologous_diff_cols, index=left_side_params.index)
    return homologous_diff_params
```
